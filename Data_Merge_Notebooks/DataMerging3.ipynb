```python
import pandas as pd
```

```python
gdp_df = pd.read_csv('./archive/GDP.csv')
data_df = pd.read_csv('./archive/Data.csv')
```

```python
# attach the GDP percent of global
glob_gdp_perc = []
gdp_df = gdp_df[gdp_df['Country Code'] == 'WLD']
for index, row in data_df.iterrows():
    NOC = row['NOC']
    year = str(row['Year'])
    ngdp = row['GDP']
    ref_df = gdp_df[gdp_df['Country Code'] == NOC]
    ggdp = gdp_df[year].iat[0]
    glob_gdp_perc.append(ngdp/ggdp)
data_df['Global_GDP%'] = glob_gdp_perc
```
