# Data Merging Part 2.

```python
# pandas
import pandas as pd
```

```python
# import the modified dataset from part 1
popped_df = pd.read_csv('./archive/AthletesWithPopulations.csv',index_col=False)

# import the datasets still needed
lifex_df = pd.read_csv('./archive/life_expectancy.csv')
ruralperc_df = pd.read_csv('./archive/RuralPopperc.csv')
gdp_df = pd.read_csv('./archive/GDP.csv')
```

```python
popped_df.head()
```

```python
# attach the rural population percentage data to the dataframe
rur_pop = []
for index, row in popped_df.iterrows():
    NOC = row['NOC']
    year = str(row['Year']) 
    ref_df = ruralperc_df[ruralperc_df['Country Code'] == NOC]
    rpop = ref_df[year].iat[0]
    rur_pop.append(rpop)
popped_df['Rural_Pop%'] = rur_pop
```

```python
# attach the life expectancy data to the dataframe
life_ex = []
for index, row in popped_df.iterrows():
    NOC = row['NOC']
    year = str(row['Year']) 
    ref_df = lifex_df[lifex_df['Country Code'] == NOC]
    lifex = ref_df[year].iat[0]
    life_ex.append(lifex)
popped_df['Life_Expect'] = life_ex
```

```python
# attach the life expectancy data to the dataframe
gdp = []
for index, row in popped_df.iterrows():
    NOC = row['NOC']
    year = str(row['Year']) 
    ref_df = gdp_df[gdp_df['Country Code'] == NOC]
    gross = ref_df[year].iat[0]
    gdp.append(gross)
popped_df['GDP'] = gdp
```

```python
data = popped_df
data.to_csv('./archive/Data.csv')
```
