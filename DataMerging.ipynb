# Olympic Data Cleaning and Merging

```python
import pandas as pd
```

```python
#pull in datasets (maybe spread out later for efficiency)
population_df = pd.read_csv('./archive/PopulationPerCountry.csv')
athletes_df = pd.read_csv('./archive/athlete_events.csv')

# the athletes_df dataset is what we will build our data around
athletes_df.head()
```

```python
# most of our indicator data begins in 1960, so cull out previous years
athletes_df = athletes_df[athletes_df['Year'] > 1959]
# sailing is not worth the trouble, cull with disdain
sailless_df = athletes_df[athletes_df['Sport'] != 'Sailing']
```

## Create a value for being from host nation

```python
#create a dict that associates the host city with the host nation's NOC
host_dict = {'Sydney':'AUS',
             'Atlanta':'USA',
             'Rio de Janeiro':'BRA',
             'Beijing':'CHN',
             'Athina':'GRC',
             'Barcelona':'ESP',
             'London':'GBR',
             'Seoul':'KOR',
             'Munich':'DEU',
             'Los Angeles':'USA',
             'Montreal':'CAN',
             'Mexico City':'MEX',
             'Roma':'ITA',
             'Tokyo':'JPN',
             'Moskva':'RUS',
             'Sochi':'RUS',
             'Vancouver':'CAN',
             'Torino':'ITA',
             'Salt Lake City':'USA',
             'Innsbruck':'AUT',
             'Nagano':'JPN',
             'Albertville':'FRA',
             'Lillehammer':'NOR',
             'Calgary':'CAN',
             'Sarajevo':'BIH',
             'Grenoble':'FRA',
             'Lake Placid':'USA',
             'Sapporo':'JPN',
             'Squaw Valley':'USA'
            }

# Fill new dataframe column for host country
host = []
for city in sailless_df['City']:
    host.append(host_dict[city])
host

sailless_df['Host'] = host

#re-empty list for filling binary values
host = []
for index, row in sailless_df.iterrows():
    # Check if the athlete is from the host country
    if row['NOC'] == row['Host']:
        # append a 1 if the athlete is from the host nation and 0 if not
        host.append(1)
    else:
        host.append(0)
# replace the Host column with the binary array
sailless_df['Host'] = host
```

## Fill Nation Population Data for Year

```python
# standardize the Olympic nation codes with ISO alpha-3 nation codes
popped_df = sailless_df
iso_codes = {population_df['Country Name'][i]: population_df['Country Code'][i] for i in range(len(population_df['Country Name']))}

new_noc = []
for index, row in sailless_df.iterrows():
    h = 0
    NOC = row['NOC']
    nation = row['Team']
    nat = nation.split('-')
    for key, value in iso_codes.items():
        if nat[0] == key:
            h = 1
            new_noc.append(value)
    if h == 0:
        if NOC =='TCH':
            new_noc.append('CZE')
        elif NOC == 'YAR' or NOC =='YMD':
            new_noc.append('YEM')
        elif NOC =='AHO':
            new_noc.append('SXM')
        elif NOC =='INA':
            new_noc.append('IDN')
        elif NOC =='SCG':
            new_noc.append('MNE')
        elif NOC =='YUG':
            new_noc.append('BIH')
        elif NOC =='MAL':
            new_noc.append('MYS')
        elif NOC =='PUR':
            new_noc.append('PRI')
        elif NOC =='CAM':
            new_noc.append('KHM')
        elif NOC =='RHO' or NOC == 'ZIM':
            new_noc.append('ZWE')
        elif NOC =='TPE':
            new_noc.append('CHN')
        elif NOC =='POR':
            new_noc.append('PRT')
        elif NOC =='MON':
            new_noc.append('MCO')
        elif NOC =='COK':
            new_noc.append('NZL')
        elif NOC =='WIF':
            new_noc.append('TTO')
        elif NOC =='DEN':
            new_noc.append('DNK')
        elif NOC =='MYA':
            new_noc.append('MMR')
        elif NOC =='BAH':
            new_noc.append('BHS')
        elif NOC =='NED':
            new_noc.append('NLD')
        elif NOC =='GER':
            new_noc.append('DEU')                  
        else:
            new_noc.append(NOC)
        
popped_df['NOC'] = new_noc

# attach the population data to the dataframe
population_df = pd.read_csv('./archive/PopulationPerCountry.csv')
nat_pop = []
glob_pop_perc = []
glob_df = population_df[population_df['Country Code'] == 'WLD']
for index, row in sailless_df.iterrows():
    NOC = row['NOC']
    year = str(row['Year']) 
    ref_df = population_df[population_df['Country Code'] == NOC]
    npop = ref_df[year].iat[0]
    gpop = glob_df[year].iat[0]
    nat_pop.append(npop)
    glob_pop_perc.append(npop/gpop)
popped_df['Nat_Pop'] = nat_pop
popped_df['Global_Pop%'] = glob_pop_perc
```
