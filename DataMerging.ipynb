```python
regions_df = pd.read_csv('./archive/noc_regions.csv')
population_df = pd.read_csv('./archive/populations.csv')
athletes_df = pd.read_csv('./archive/athlete_events.csv')
population_df.head()
```

```python
athletes_df.head()
```

```python
athletes_df.set_index(['NOC','Year'],inplace=True)
athletes_df.head()
```

```python
population_df.set_index(['NOC','Year'],inplace=True)
population_df.head()
```

```python
popped_raw = athletes_df.join(population_df,on=['NOC','Year'])
popped_df = popped_raw.dropna(subset=['Value'])
popped_df.head()
```
